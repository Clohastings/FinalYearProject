```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
# this command is necessary to view the plots directly in Jupyter Notebooks

#creating x coordinates
x_axis = np.linspace(0,5)

#creating y coordinate
def line(m):
    y = m*x_axis
    return y
#plotting line
plt.plot(x_axis, line(3))
```
